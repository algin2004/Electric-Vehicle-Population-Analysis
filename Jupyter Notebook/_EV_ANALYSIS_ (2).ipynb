```python
# WHAT IS THE AVERAGE AND MEDIAN ELECTRIC RANGE ACROSS ALL VEHICLE
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("Electric_Vehicle_Population_Data.csv")
avg_range=df["Electric Range"].mean()
median_range = df["Electric Range"].median()
print(f"The Average Electric Range across all vechicles :{avg_range:.2f}")
print(f"The Median Electric Range across all vechiles :{median_range:.2f}")
plt.figure(figsize=(12,6))
sns.histplot(data=df,x="Electric Range",bins = 13,kde=True,color="green")
plt.axvline(avg_range,color = 'red',linestyle='--',linewidth = 2,label = f"Mean :{avg_range:.1f}")
plt.axvline(median_range,color ='blue',linestyle='--',linewidth=2,label=f"Median :{median_range:.1f}")
plt.title("Electric Range Distribution")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# HOW DOES ELECTRIC RANGE VARIES ACROSS DIFFERENT MAKES AND MODELS
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("Electric_Vehicle_Population_Data.csv")
df['make_model']=df['Make']+" - "+df['Model']

avg_range = df.groupby('make_model')["Electric Range"].mean().reset_index()
avg_evrange = df['Electric Range'].mean()
top_make_model = avg_range.sort_values(by="Electric Range",ascending=False).head(10)
plt.figure(figsize=(15,6))
sns.barplot(data=top_make_model,x="make_model",y="Electric Range",label=f"Average Electric Range:{avg_evrange:1f}")
plt.title("electric range across different make and models")
plt.xticks(rotation = 45)
plt.legend()
plt.show()
                       
```

```python
# GROWTH OF ELECTRIC VEHICLES BY YEAR
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("Electric_Vehicle_Population_Data.csv")
counts = df.groupby(['Model Year','Electric Vehicle Type']).size().reset_index(name='Count')
plt.figure(figsize=(10,6))
sns.lineplot(data=counts , x = 'Model Year', y='Count',hue="Electric Vehicle Type")
plt.xlabel("Model Year")
plt.ylabel("Number of Registered Vehicles")
plt.title("Growth Of EV By Years")
plt.legend(title="EV Type")
plt.tight_layout()
plt.show()
```

```python
# HOW HAS NUMBER OF EVs PER STATE CHANGED OVER TIME
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("Electric_Vehicle_Population_Data.csv")
state_year = df.groupby(['Model Year','State']).size().reset_index(name='counts')
state_top = df['State'].value_counts().head(10).index
state_year_top = state_year[state_year['State'].isin(state_top)]
plt.figure(figsize=(15,6))
sns.lineplot(data=state_year_top,x='Model Year',y='counts',hue='State',marker='o')
plt.title("No Of EV per State Over Years")
plt.ylabel("No Of EV's")
plt.grid(True)
plt.show()
```

```python
# PREMIUM VS BUDGET CATAGORIES
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("Electric_Vehicle_Population_Data.csv")
avg_price=df['Base MSRP'].mean()
print(avg_price)
df['price_range']=np.where(df['Base MSRP']<=727,"Budget","Premium")
product_range=df['price_range'].value_counts().reset_index()
print(product_range)
plt.figure(figsize=(15,6))
sns.barplot(data=product_range,x='price_range',y='count',color='orange',edgecolor='black')
plt.title("PREMIUM Vs BUDGET CATEGORY")
plt.show()
```

```python
# TOP 10 MAKES BY PREMUM AND BUDGET CATAGORY
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("Electric_Vehicle_Population_Data.csv")
avg_price=df['Base MSRP'].mean()
print(avg_price)
df['price_range']=np.where(df['Base MSRP']<=727,"Budget","Premium")
product_range=df['price_range'].value_counts().reset_index()
print(product_range)
top_makes=df['Make'].value_counts().head(10).index
df_top_makes =df[df['Make'].isin(top_makes)]
plt.figure(figsize=(15,6))
sns.countplot(data=df_top_makes,x='Make',hue=df['price_range'])
plt.title("top 10 makes with premium and budget catagories")
plt.show()
```
